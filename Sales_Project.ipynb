```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel(r"C:\Users\kunal dhingra\Desktop\Sales Regression.xlsx")
data
```

```python
import sklearn
from sklearn import linear_model
```

```python
x= data.drop('Sales generated per dollar spent',axis = 'columns')
x
```

```python
y= data.drop('Cost for Advertisement', axis = 'columns')
y
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
a=((reg.score(x,y))*100)
b = round(a,2)
print(f'The efficiency of the model is {b}%')
```
